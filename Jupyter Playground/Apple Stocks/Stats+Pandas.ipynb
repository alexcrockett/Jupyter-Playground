```python
import numpy as np
import pandas as pd
apples = pd.read_csv('AppleStockPrices.csv')
```

```python
# Calculating means and storing variables for later
mean_Open = apples['Open'].mean() # Mean Open
mean_High = apples['High'].mean() # Mean High
mean_Low = apples['Low'].mean() # Mean Low
mean_Close = apples['Close'].mean() # Mean Close
mean_Volume = apples['Volume'].mean()


# Calculating the mean values and storing them in a dictionary
mean_values = {
    "Mean Open": apples['Open'].mean(), # Mean Open
    "Mean High": apples['High'].mean(), # Mean High
    "Mean Low": apples['Low'].mean(), # Mean Low
    "Mean Close": apples['Close'].mean(), # Mean Close
    "Mean Volume": apples['Volume'].mean() # Mean Volume
}

# Iterating over the dictionary and printing each item
for key, value in mean_values.items():
    print(f"{key}: {value:.2f}")
```

```python
# Create an array of values that can be manipulated later

mean_array = x_1 = np.array([mean_Open, mean_High, mean_Low, mean_Close, mean_Volume])

print(x_1)
```

```python
# Here we want to convert our array into a column vector
x_1 = mean_array.reshape((-1, 1))
```

```python
print(x_1)
```

```python
# Lets make a very simple plot

import matplotlib.pyplot as plt # Import the library we want to use
```

```python
# Now we have the package install and imported let's make a plot

## creating the dataset with the dictionary definition earlier
import matplotlib.pyplot as plt

# Extracting keys and values
labels = list(mean_values.keys())
means = list(mean_values.values())

# Creating a bar chart
plt.bar(labels, means)
plt.ylabel('Mean Values')
plt.title('Mean Values of Apple Stock Characteristics')
plt.xticks(rotation=45)  # Rotates labels to make them readable
plt.show()
```

```python
# using the array
# Labels corresponding to the order of means in mean_array
labels = ["Mean Open", "Mean High", "Mean Low", "Mean Close", "Mean Volume"]

# Creating a bar chart
plt.bar(labels, mean_array)  # mean_array is already in the correct format
plt.ylabel('Mean Values')
plt.title('Mean Values of Apple Stock Characteristics')
plt.xticks(rotation=45)  # Rotates labels to make them readable
plt.show()
```

```python
# Ommiting Volume with the dictionary
# Using dictionary comprehension to exclude 'Mean Volume'
filtered_mean_values = {k: v for k, v in mean_values.items() if k != 'Mean Volume'}

# Extracting keys and values
labels = list(filtered_mean_values.keys())
means = list(filtered_mean_values.values())

# Plotting the bar chart
plt.bar(labels, means)
plt.ylabel('Mean Values')
plt.title('Filtered Mean Values of Apple Stock Characteristics')
plt.xticks(rotation=45)
plt.show()
```

```python
# Omitting volume with the array
# Labels corresponding to the order of means, excluding 'Mean Volume'
labels = ["Mean Open", "Mean High", "Mean Low", "Mean Close"]

# Creating a bar chart (excluding the last element of mean_array)
plt.bar(labels, mean_array[:-1])  # Excludes the last element (Mean Volume)
plt.ylabel('Mean Values')
plt.title('Filtered Mean Values of Apple Stock Characteristics')
plt.xticks(rotation=45)
plt.show()
```
